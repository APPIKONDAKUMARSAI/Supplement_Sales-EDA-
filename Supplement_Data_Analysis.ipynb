```python
# Revenue by Platform
sns.barplot(data=df, x='Platform', y='Revenue', estimator=sum)
plt.title("Total Revenue by Platform")
plt.ylabel("Revenue")
plt.show()
print("Insight: Some platforms clearly outperform others in total revenue, indicating strong channel preference.")
```

```python
# Return rate by Platform
sns.set_style("white")
sns.boxplot(data=df, x='Platform', y='Return Rate')
plt.title("Return Rate Distribution by Platform")
plt.show()
print("Insight: Some platforms exhibit significantly higher return rates, suggesting quality or service concerns.")
```

```python
# Revenue by Category
plt.figure()
df.groupby('Category')['Revenue'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title("Revenue by Category")
plt.ylabel("Total Revenue")
plt.show()
print("Insight: A few categories dominate revenue generation; these may deserve focused marketing efforts.")
```

```python
sns.heatmap(df[['Units Sold', 'Units Returned', 'Price', 'Discount', 'Return Rate', 'Revenue']].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
print("Insight: Units Sold is positively correlated with Revenue and inversely correlated with Price and Discount.")
```

```python
# Insights
# 1. Product with highest total revenue
top_revenue_product = df.groupby('Product Name')['Revenue'].sum().idxmax()
print("\nProduct with Highest Revenue:", top_revenue_product)

# idxmax()
# This finds the product with the highest total revenue by identifying the index (or label) of the maximum value in the summed revenue series.
```

```python
# 2. Top 5 best-selling products by units sold
top_5_units_sold = df.groupby('Product Name')['Units Sold'].sum().sort_values(ascending=False).head(5)

print("\nTop 5 Best-Selling Products by Units Sold:\n", top_5_units_sold)
```

```python
# 3. Category with highest average revenue per product
category_avg_revenue = df.groupby('Category')['Revenue'].mean().sort_values(ascending=False)

print("\nCategory with Highest Average Revenue per Product:\n", category_avg_revenue.head(1))
```

```python
# 4. Platform with most sales (units and revenue)
platform_units = df.groupby('Platform')['Units Sold'].sum().idxmax()
platform_revenue = df.groupby('Platform')['Revenue'].sum().idxmax()
print("\nPlatform with Most Units Sold:", platform_units)
print("Platform with Most Revenue:", platform_revenue)
```

```python
# 5. Location with consistently highest revenue
top_location = df.groupby('Location')['Revenue'].sum().idxmax()

print("\nLocation with Highest Total Revenue:", top_location)
```
